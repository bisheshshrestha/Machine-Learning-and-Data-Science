```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data_Sets/traffic_accidents.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
numerical_column = df.select_dtypes(include=['number','int64'])
```

```python
numerical_column.head()
```

```python
corr_matrix = numerical_column.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidth=0.5)

plt.title('Heatmap traffic accident')
plt.show()
```
